# exemplo 02 - Atribuição de variaveis

```python
# tudo é tratado como objeto (armazenar duas coisas - o que é e valor)

carros = 100

# objeto = numero inteiro, e valor = 100 - rotulo --> nome da variável [cars]

espacoNoCarro = 4.0 # real ou ponto flutuante

motoristas = 30

passageiros = 90

carrosSemMotorista = carros - motoristas

carrosComMotorista = motoristas

capacidadeDePassageiros = carrosComMotorista * espacoNoCarro

mediaPassageirosPorCarro = passageiros / carrosComMotorista

print("Existem", carros, "disponiveis.")

print(f"Existem apenas {motoristas} disponiveis.")

print('Nos podemos transportar {} pessoas hoje'.format(capacidadeDePassageiros))

print(100/5)

# mais um exemplo de uso do format

print("{} + {} + {} + {}".format(mediaPassageirosPorCarro, capacidadeDePassageiros, carrosComMotorista, carrosSemMotorista))

# exemplo com print

print(f"{mediaPassageirosPorCarro}, {capacidadeDePassageiros}, {carrosComMotorista}, {carrosSemMotorista}")

# as operações moldam o tipo de dado que vai ser apresentado

# int / int = real
```

```python
i = 0

# variavel simples - armazena apenas um valor

# variavel composta - armazena vários valores

# Lista - mutáveis: [a, b, c, a, b, [ad]]

# Tupla - imutáveis: (a, b, c, 45, asd, asd, (asdfscv))

# Conjuntos - mutáveis mas sem repetições: {valor1, valor2, valor3, ...}

numero = [] # inicializar uma variavel composta do tipo lista

# WHILE

while(i < 6):
  print(f"Estou indo de 0 até {i}")
  numero.append(i)
  i = i + 1
  print(f'O valor da lista é {numero}')
  # aqui estou dentro da repetição

# aqui estou fora da repetição

# FOR

for i in range(0,6):
  print(f"Estou indo de 0 até {i}")
  numero.append(i)
  print(f"O valor da lista é {numero}")


print(f'\n\n Vamos experimentar o FOR com strings.\n\n')

texto = "Aula de sexta feira"

for i in texto:
  print(f"O contaor está com a letra {i}")


```
