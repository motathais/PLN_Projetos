1 - Elaborar um programa que calcule e apresente o valor do volume de uma caixa retangular, utilizando a fórmula VOLUME = COMPRIMENTO * LARGURA * ALTURA

```python
comprimento = float(input("Informe o comprimento (m): "))

largura = float(input("Informe a largura (m): "))

altura = float(input("Informe a altura em (m): "))

volume = comprimento * largura * altura

print(f" O volume é: {volume}")
```

3 - Buscar e substituir palavras em um texto;

```python
texto = input("Informe um texto: ").split()

palavra = input("Informe a palavra que quer encontrar: ")

if palavra in texto:

  ocorrencia = texto.count(palavra)

  print(f"A palavra aparece {ocorrencia} vezes." )

  opcao = input("Caso deseje substituir a palavra digite 1, para encerrar digite qualquer caractere: ")

  if opcao == "1":

    substituta = input("Por qual palavra gostaria de substituir?: ")

    texto = [substituta if p == palavra else p for p in texto]

    print(f"O texto ficou assim: {' '.join(texto)}")

  else:
    print("Programa encerrado!")
    exit()

else:
  print("Palavra não encontrada!")
  exit()




```

4 - Separar frases usando split() e iterar sobre elas.

```python
texto = input("Informe um texto: ")

frases = texto.replace("?", ".").replace("!", ".").replace(",", ".").split(".")

for frase in frases:
  print(frase.strip())

```

5 - Ler três valores inteiros representados pelas variáveis A, B e C e apresentar os valores lidos dispostos em ordem crescente.

```python
a = int(input("Informe o valor de A: "))
b = int(input("Informe o valor de B: "))
c = int(input("Informe o valor de C: "))

numeros = [a, b, c]

numeros.sort()

print("Numeros em ordem crescente:", numeros)


```
